## Part 1 -  Artificial Intelligence (AI) vs Machine Learning (ML)

We start by clarifying a common source of confusion: **AI and ML are not the same thing**.

**Artificial Intelligence (AI)** is the *umbrella concept*.  
It refers to systems designed to perform tasks that normally require human intelligence, such as:
- reasoning
- perception
- language understanding
- decision-making

**Machine Learning (ML)** is a *subset of AI*.  
It refers specifically to systems that **learn patterns from data** in order to make predictions or decisions, without being explicitly programmed with rules.

### A practical analogy

Think of **AI as the organisation**, and **ML as the engines inside it**.

- **AI** is the *overall capability*  
  - The full system that delivers intelligent behaviour
  - Often combines rules, optimisation, ML models, heuristics, and human inputs

- **ML** is the *learning component*  
  - Models trained on historical data
  - Used to predict, rank, classify, or forecast

A real-world example:

- **Google as a company** builds *AI systems*  
  - Search, Maps, Ads, Assistant, Gmail, YouTube
- **Machine Learning models** power *specific functions* inside those systems  
  - Ranking search results  
  - Recommending YouTube videos  
  - Filtering spam in Gmail  

ML does not replace the system but it **supports decisions within the system**.

<br>

### How this maps to finance

In finance:
- **AI systems** include:
  - trading platforms
  - risk management systems
  - portfolio construction workflows
  - reporting and decision-support tools

- **ML models** sit inside these systems to:
  - forecast returns or volatility
  - classify risk states
  - rank assets
  - detect anomalies or signals

An ML model on its own is **not a trading system**.  
It becomes useful only when embedded inside:
- data pipelines
- evaluation logic
- risk controls
- execution constraints
- reporting and governance layers

## Part 2 - Predictive Modelling vs Causal Inference

A critical distinction in financial machine learning is the difference between **prediction** and **causation**.

### Predictive modelling
**Predictive modelling** focuses on learning patterns from historical data in order to estimate what is *likely* to happen next.

<p align="center">
<img src="https://images.unsplash.com/photo-1579019163248-e7761241d85a?q=80&w=870&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D" width="300" height="300">
</p>

*Credits:* Unsplash, Anastase Maragos

- *“Given what has happened before, what is the most probable outcome?”*
- It relies on correlations, patterns, and statistical regularities
- It does not require understanding the underlying economic mechanism


### Causal inference

**Causal inference** focuses on understanding cause-and-effect relationships.

<p align="center">
<img src="https://images.unsplash.com/photo-1669232696081-6f186ac7d4e4?q=80&w=1332&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D" width="300" height="300">
</p>

*Credits:* Unsplash, Delano Ramdas


- *“If we change something, what will happen as a result?”*
- It requires strong assumptions, controlled experiments, or natural experiments
- It is significantly harder to do reliably in financial markets

<br>

The following table summarises the differences between predictive and causal in several different scenarios:

| Scenario | Predictive Modelling (What is likely to happen?) | Causal Inference (What happens if we intervene?) |
|--------|-----------------------------------------------|------------------------------------------------|
| Asset prices | Forecast short-term returns or volatility using historical price patterns | Estimate whether an interest rate change *causes* asset prices to move |
| Credit risk | Predict probability of default based on borrower characteristics | Assess whether changing lending criteria *reduces* default risk |
| Portfolio allocation | Rank assets by expected performance | Determine whether reallocating capital *causes* improved portfolio outcomes |
| Trading strategies | Predict when a strategy is likely to perform well | Measure the true impact of introducing or modifying a strategy rule |

### Example: When Models Learn Patterns, Not Causes

**Claim:** Increasing the market signal causes asset returns to increase.

This claim sounds reasonable because the market signal and returns move together in historical data but we will show that this conclusion is wrong.

In other words, a model can be **predictively useful** while being **causally misleading**.

<br>

#### **Setup 1**

- We observe a market signal (`signal`) that appears predictive of returns (`returns`)
- In reality, both the signal and returns are driven by an unobserved factor (`macro`)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

np.random.seed(42)
n = 300

# Unobserved macro factor (e.g., risk sentiment / liquidity)
macro = np.random.normal(0, 1, n)

# Observable signal (e.g., technical indicator) correlated with macro
signal = macro + np.random.normal(0, 0.3, n)

# Returns driven by macro (not signal)
returns = 1.5 * macro + np.random.normal(0, 0.3, n)

df = pd.DataFrame({"signal": signal, "returns": returns})

# Fit predictive model: returns ~ signal
model = LinearRegression().fit(df[["signal"]], df["returns"])
x_line = np.linspace(df["signal"].min(), df["signal"].max(), 100)
y_line = model.predict(x_line.reshape(-1, 1))

plt.figure()
plt.scatter(df["signal"], df["returns"])
plt.plot(x_line, y_line)
plt.title("Market Signal vs Returns")
plt.xlabel("Market Signal")
plt.ylabel("Returns")
plt.show()
```

```python
corr_table = df[['signal', 'returns']].corr().round(3)
corr_table
```

The correlation between `signal` and `returns` is very high (≈ 0.94), as shown in the table above.

This tells us:
- The two variables move together in historical data
- The signal is **predictively useful**

It does **not** tell us:
- That the signal causes returns
- That changing the signal will change outcomes

In this example:
- Both variables are driven by the same underlying factor
- Correlation reflects shared information, not control

<br>

### **Setup 2**

- We observe both the market signal (`signal`) and returns (`returns`) over time  
- Both series move together because they are influenced by the same unobserved factor (`macro`)  
- The signal appears informative because it reflects underlying market conditions  
- At this stage, we are still **observing**, not intervening  

This setup helps explain *why* the predictive relationship exists,  without yet claiming that the signal causes returns.


```python
plt.figure(figsize=(12, 6))
plt.plot(df['signal'].values, label='Market Signal', color='blue')
plt.plot(df['returns'].values, label='Returns', color='orange')
plt.title('Signal and Returns over Time')
plt.xlabel('Observation')
plt.ylabel('Value')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

**Key observations**

- The market signal and returns move together over time  
- Periods of high and low values tend to align  
- Co-movement suggests a shared underlying driver  
- This explains why the signal is predictive, not why it is causal  


<br>

### **Setup 3**

- We intervene by **forcing the market signal** to a constant value  
- This simulates a direct attempt to control the signal  
- The underlying driver (`macro`) is unchanged  
- Returns continue to be driven by the macro factor, not the signal  
- If the signal were causal, returns would respond to this intervention  
- If the signal is only predictive, returns will remain unaffected  


```python
# "Intervene" on the signal: force it to a constant (do(signal = constant))
forced_signal = np.zeros(n)  # intervention: signal is set externally

plt.figure()
plt.scatter(forced_signal, returns)
plt.title("Intervention: Forcing the Signal Does NOT Change Returns")
plt.xlabel("Forced Market Signal (constant)")
plt.ylabel("Returns")
plt.show()
```

What is observed:
- All points align vertically because the signal no longer varies  
- Returns still vary widely despite the signal being fixed  
- There is no systematic change in returns when the signal is controlled  

What this tells us:
- Changing the signal does **not** change returns  
- The earlier relationship relied on correlation, not causation  
- The signal is not a lever that controls outcomes
